```python
import pandas as pd
import json

# 读取 XLSX 文件
file_path = 'datasets/AI.xlsx'
df = pd.read_excel(file_path) #, nrows=100
df.drop(columns=['資料來源'], inplace=True)
df['正確答案'] = df['正確答案'].replace({'選項1': 1, '選項2': 2, '選項3': 3, '選項4': 4})
df
```

```python
# 转换为 JSON
json_list = []

for index, row in df.iterrows():
    json_obj = {
        # "id": "{}".format(row['ID']),
        "instruction": "請根據以下文章，以數字回答選擇題:\n",
        "input": "{}\n\n{}\n1: {}\n2: {}\n3: {}\n4: {}\n".format(row['文章'], row['問題'], row['選項1'], row['選項2'], row['選項3'], row['選項4']),
        "output": str(row['正確答案'])
    }
    json_list.append(json_obj)

# 保存 JSON 文件
with open('datasets/AI.json', 'w', encoding='utf-8') as json_file:
    json.dump(json_list, json_file, ensure_ascii=False, indent=4)
```

```python
import pandas as pd

# 读取 Excel 文件
file_path = 'datasets/AI.xlsx'  # 替换为您的文件路径
df = pd.read_excel(file_path)
df.drop(columns=['資料來源'], inplace=True)
df['正確答案'] = df['正確答案'].replace({'選項1': 1, '選項2': 2, '選項3': 3, '選項4': 4})

# 检查 DataFrame 中的行数
num_rows = df.shape[0]

# 如果 DataFrame 至少有100行，随机选择100行；否则选择所有行
if num_rows >= 100:
    df_sample = df.sample(n=100)
else:
    df_sample = df

# 显示随机挑选的结果
print(df_sample)
```

```python
# 转换为 JSON
json_list = []

for index, row in df_sample.iterrows():
    json_obj = {
        # "id": "{}".format(row['ID']),
        "instruction": "請根據以下文章，以數字回答選擇題:\n",
        "input": "{}\n\n{}\n1: {}\n2: {}\n3: {}\n4: {}\n".format(row['文章'], row['問題'], row['選項1'], row['選項2'], row['選項3'], row['選項4']),
        "output": str(row['正確答案'])
    }
    json_list.append(json_obj)

# 保存 JSON 文件
with open('datasets/valid.json', 'w', encoding='utf-8') as json_file:
    json.dump(json_list, json_file, ensure_ascii=False, indent=4)
```
